```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from tensorflow import keras
from tensorflow.keras import layers
```

```python
# Replace 'ETP_Dataset.csv' with the actual path and filename of your CSV file
data = pd.read_csv('/content/drive/MyDrive/BUET Thesis/ETP_Dataset.csv')
```

```python
# Split the data into features (X) and the target variable (y)
X = data.drop(columns=['ETP quotidien [mm]', 'Date/heure'])
y = data['ETP quotidien [mm]']

# Normalize the features to scale the data to a range [0, 1]
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
model = keras.Sequential([
    layers.Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    layers.Dense(32, activation='relu'),
    layers.Dense(1)  # Output layer with 1 neuron for regression
])

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
model.fit(X_train, y_train, epochs=100, batch_size=32, validation_split=0.2)
```

```python
loss = model.evaluate(X_test, y_test)
print("Test loss:", loss)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from tensorflow import keras
from tensorflow.keras import layers
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
# Replace 'ETP_Dataset.csv' with the actual path and filename of your CSV file
data = pd.read_csv('/content/drive/MyDrive/BUET Thesis/ETP_Dataset.csv')
```

```python
# Split the data into features (X) and the target variable (y)
X = data.drop(columns=['ETP quotidien [mm]', 'Date/heure'])
y = data['ETP quotidien [mm]']

# Normalize the features to scale the data to a range [0, 1]
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
ann_model = keras.Sequential([
    layers.Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    layers.Dense(32, activation='relu'),
    layers.Dense(1)  # Output layer with 1 neuron for regression
])

# Compile the ANN model
ann_model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
ann_model.fit(X_train, y_train, epochs=100, batch_size=32, validation_split=0.2)
```

```python
# Fit ARIMA model
arima_model = sm.tsa.ARIMA(y_train, order=(1, 0, 1))
arima_model_fit = arima_model.fit()
```

```python
# Predict using ANN model
ann_predictions = ann_model.predict(X_test).flatten()


```

```python
# Plot actual vs. predicted values using ANN model
plt.figure(figsize=(10, 6))
plt.plot(y_test.values, label='Actual')
plt.plot(ann_predictions, label='ANN Predictions')
plt.legend()
plt.title('Actual vs. ANN Predictions')
plt.xlabel('Time')
plt.ylabel('ETP quotidien [mm]')
plt.show()
```

```python
# Predict using ARIMA model
forecast_length = len(y_test)
arima_predictions = arima_model_fit.forecast(steps=forecast_length)

# Create a new index that aligns with the test data index
arima_index = pd.date_range(start=y_test.index[-1], periods=forecast_length, closed='right')

# Convert the ARIMA predictions to a Pandas Series with the new index
arima_predictions_series = pd.Series(arima_predictions, index=arima_index)
```

```python
# Plot actual vs. predicted values using ARIMA model
plt.figure(figsize=(10, 6))
plt.plot(y_test, label='Actual')
plt.plot(arima_predictions_series, label='ARIMA Predictions')
plt.legend()
plt.title('Actual vs. ARIMA Predictions')
plt.xlabel('Time')
plt.ylabel('ETP quotidien [mm]')
plt.show()
```

The relationship between Time and ETP quotidien [mm] (ETP stands for Evapotranspiration) typically refers to the change in evapotranspiration over time. Evapotranspiration is the combined process of water evaporation from the Earth's surface (mainly soil and water bodies) and transpiration from plants. It represents the amount of water that is lost from the Earth's surface due to these processes.

The relationship between Time and ETP quotidien [mm] can vary depending on various factors such as climate, geographical location, vegetation, temperature, humidity, and precipitation. Some key points to consider about this relationship are:

    Seasonality: ETP tends to show seasonal patterns due to changes in temperature, solar radiation, and precipitation throughout the year. In many regions, ETP is higher during warmer months when temperatures are high, and solar radiation is intense.

    Diurnal Variations: ETP can also exhibit diurnal variations, especially in areas with significant temperature fluctuations between day and night.

    Climate Zones: Different climate zones (e.g., tropical, temperate, arid, etc.) will have distinct patterns of ETP over time.

    Impact of Precipitation: Precipitation can affect ETP, especially in arid regions where increased rainfall can result in a higher ETP due to increased evaporation.

    Vegetation Cover: Areas with dense vegetation will have higher transpiration rates, leading to potentially higher ETP values.

    Water Availability: In areas with limited water availability, the ETP might be an essential factor in managing water resources.

Understanding the relationship between Time and ETP quotidien [mm] is crucial for various applications, including agricultural water management, hydrology, climate modeling, and water resource planning.
